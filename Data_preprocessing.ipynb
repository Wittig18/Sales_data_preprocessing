Perform a hyperparameter tuning for `RandomForestRegressor` on the prepared training data (`x_train`, `y_train`) using `GridSearchCV`. Define a parameter grid that includes values for `n_estimators`, `max_depth`, `min_samples_split`, and `min_samples_leaf`. After fitting the `GridSearchCV` object, evaluate the `best_estimator_` on the test set (`x_test`, `y_test`) and print its R2 score.

## Define Parameter Grid

### Subtask:
Define a dictionary of hyperparameters and their possible values for the RandomForestRegressor to search through.


**Reasoning**:
The subtask requires defining a parameter grid for hyperparameter tuning. I will create a dictionary with the specified hyperparameters and their possible values.



```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
print("Parameter grid defined successfully.")
```

**Reasoning**:
The next logical step is to import the `GridSearchCV` module to perform hyperparameter tuning for the `RandomForestRegressor`.



```python
from sklearn.model_selection import GridSearchCV
print("GridSearchCV imported successfully.")
```

**Reasoning**:
Now that the parameter grid is defined and `GridSearchCV` is imported, I will initialize the `GridSearchCV` object with the `RandomForestRegressor` model and the defined `param_grid`. This sets up the search process for optimal hyperparameters.



```python
grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42), param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
print("GridSearchCV initialized.")
```

**Reasoning**:
With the `GridSearchCV` object initialized, the next step is to fit it to the training data (`x_train`, `y_train`) to perform the hyperparameter search and find the best model.



```python
grid_search.fit(x_train, y_train)
print("GridSearchCV fitting complete.")
```

**Reasoning**:
After fitting `GridSearchCV`, I will retrieve the best estimator found during the search, which is the RandomForestRegressor model with the optimal hyperparameters.



```python
best_rf_model = grid_search.best_estimator_
print("Best RandomForestRegressor model retrieved.")
```

**Reasoning**:
Now that the best model has been retrieved, I will use it to make predictions on the test set (`x_test`) to evaluate its performance.



```python
y_pred_best_rf = best_rf_model.predict(x_test)
print("Predictions made using the best RandomForestRegressor model.")
```

**Reasoning**:
To evaluate the performance of the best model, I will calculate its R2 score using the actual values (`y_test`) and the predicted values (`y_pred_best_rf`).



```python
r2_best_rf = r2_score(y_test, y_pred_best_rf)
print(f"R2 score for the best RandomForestRegressor model: {r2_best_rf}")
```
